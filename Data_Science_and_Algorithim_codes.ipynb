```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

# **Linear Regression**

```python
Housing_data_df=pd.read_csv("Housing.csv")
Housing_data_df.head()
```

```python
Housing_data_df = pd.get_dummies(Housing_data_df, columns=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus'], drop_first=True)
Housing_data_df = Housing_data_df.astype(int)
Housing_data_df.head()
```

```python
Housing_data_df['area']=(Housing_data_df['area']-Housing_data_df['area'].mean())/Housing_data_df['area'].std()
Housing_data_df.head()
```

```python
X=Housing_data_df[['area', 'bedrooms','bathrooms','stories','parking','mainroad_yes','guestroom_yes','basement_yes','hotwaterheating_yes','airconditioning_yes','prefarea_yes','furnishingstatus_semi-furnished','furnishingstatus_unfurnished']]
Y=Housing_data_df['price']
```

```python
LRM=LinearRegression()
```

```python
X_Train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=20)
```

```python
LRM.fit(X_Train,y_train)
```

```python
Y_pred=LRM.predict(X_test)
```

```python
mean_squared_error(y_test,Y_pred)
```

```python
r2_score(y_test,Y_pred)
```

```python
LRM.coef_
```

```python
LRM.intercept_
```

# **Polynomial Regression**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
X = Housing_data_df.drop(columns=['price'])  # Replace 'price' with the actual target column
y = Housing_data_df['price']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
degree = 1  # You can experiment with different degrees
poly = PolynomialFeatures(degree=degree)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_poly, y_train)
```

```python
y_train_pred = model.predict(X_train_poly)
y_test_pred = model.predict(X_test_poly)
```

```python
test_mse = mean_squared_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)

print(f"Test MSE: {test_mse}")
print(f"Test R²: {test_r2}")
```
